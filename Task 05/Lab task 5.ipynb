```python
def pixelVal(pix, r1, s1, r2, s2):
    if (0 <= pix and pix <= r1):
        return (s1 / r1)*pix
    elif (r1 < pix and pix <= r2):
        return ((s2 - s1)/(r2 - r1)) * (pix - r1) + s1
    else:
        return ((255 - s2)/(255 - r2)) * (pix - r2) + s2

img = cv2.imread('sample.jpg')

r1 = 70
s1 = 0
r2 = 140
s2 = 255

pixelVal_vec = np.vectorize(pixelVal)

contrast_stretched = pixelVal_vec(img, r1, s1, r2, s2)

cv2.imwrite('contrast_stretch.jpg', contrast_stretched)
```

```python
image = cv2.imread('sample.jpg')
 
mask = np.zeros(image.shape[:2], np.uint8)
 
backgroundModel = np.zeros((1, 65), np.float64)
foregroundModel = np.zeros((1, 65), np.float64)
 
rectangle = (20, 100, 150, 150)
 
cv2.grabCut(image, mask, rectangle,  
            backgroundModel, foregroundModel,
            3, cv2.GC_INIT_WITH_RECT)
 
mask2 = np.where((mask == 2)|(mask == 0), 0, 1).astype('uint8')
 
image_segmented = image * mask2[:, :, np.newaxis]
 
plt.subplot(1, 2, 1)
plt.title('Original Image')
plt.imshow(cv2.cvtColor(image, cv2.COLOR_BGR2RGB))
plt.axis('off')

plt.subplot(1, 2, 2)
plt.title('Segmented Image')
plt.imshow(cv2.cvtColor(image_segmented, cv2.COLOR_BGR2RGB))
plt.axis('off')

plt.show()
```

```python
img = cv2.imread('sample.jpg')
gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
edges = cv2.Canny(gray, 50, 150, apertureSize=3)

plt.figure(figsize=(12, 5))
plt.subplot(1, 2, 1)
plt.imshow(cv2.cvtColor(img, cv2.COLOR_BGR2RGB))
plt.title('Original Image')
plt.axis('off')

plt.subplot(1, 2, 2)
plt.imshow(edges, cmap='gray')
plt.title('Detected Edges')
plt.axis('off')
plt.tight_layout()
plt.show()

lines = cv2.HoughLines(edges, 1, np.pi/180, 100)

if lines is not None:
    img_lines = img.copy()
    
    for r_theta in lines:
        arr = np.array(r_theta[0], dtype=np.float64)
        r, theta = arr
        a = np.cos(theta)
        b = np.sin(theta)
        x0 = a*r
        y0 = b*r
        x1 = int(x0 + 1000*(-b))
        y1 = int(y0 + 1000*(a))
        x2 = int(x0 - 1000*(-b))
        y2 = int(y0 - 1000*(a))
        cv2.line(img_lines, (x1, y1), (x2, y2), (0, 0, 255), 2)

    cv2.imwrite('linesDetected.jpg', img_lines)
    
    plt.figure(figsize=(12, 5))
    plt.imshow(cv2.cvtColor(img_lines, cv2.COLOR_BGR2RGB))
    plt.title(f'Lines Detected ({len(lines)} lines found)')
    plt.axis('off')
    plt.tight_layout()
    plt.show()
    
    print(f"Lines detected and saved as 'linesDetected.jpg' - Found {len(lines)} lines")
else:
    print("No lines detected in the image - try adjusting edge detection or Hough parameters")
    print("Suggestions: Lower Canny thresholds or reduce Hough threshold value")
```
