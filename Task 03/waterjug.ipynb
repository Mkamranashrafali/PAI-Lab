# WaterJug Problem

```python
class JugSolver:
    
    def __init__(self, capacitya, capacityb, goal):
        self.capacitya = capacitya
        self.capacityb = capacityb
        self.goal = goal
        self.explored = set()
        self.path = []
    
    def check_target(self, state):
        return state[0] == self.goal or state[1] == self.goal
    
    def generatemoves(self, current):
        a, b = current
        possible_moves = []
        
        if a < self.capacitya:
            possible_moves.append(((self.capacitya, b), f"Fill first jug completely"))
        
        if b < self.capacityb:
            possible_moves.append(((a, self.capacityb), f"Fill second jug completely"))
        
        if a > 0:
            possible_moves.append(((0, b), f"Empty first jug"))
        
        if b > 0:
            possible_moves.append(((a, 0), f"Empty second jug"))
        
        if a > 0 and b < self.capacityb:
            transfer = min(a, self.capacityb - b)
            new_a = a - transfer
            new_b = b + transfer
            possible_moves.append(((new_a, new_b), f"Pour {transfer} liters from first jug to second jug"))
        
        if b > 0 and a < self.capacitya:
            transfer = min(b, self.capacitya - a)
            new_a = a + transfer
            new_b = b - transfer
            possible_moves.append(((new_a, new_b), f"Pour {transfer} liters from second jug to first jug"))
        
        return possible_moves
    
    def search(self, state):
        if state in self.explored:
            return False
        
        self.explored.add(state)
        
        if self.check_target(state):
            return True
        
        for next_state, move in self.generatemoves(state):
            if self.search(next_state):
                self.path.append(move)
                return True
        
        return False
    
    def run(self):
        start = (0, 0)
        print(f"Initial state: Both jugs are empty")
        print(f"Goal: Get {self.goal} liters in any jug")
        print(f"Jug capacities: {self.capacitya}L and {self.capacityb}L")
        print()
        
        if self.search(start):
            print("Solution found! Here are the steps:")
            print()
            for i, step in enumerate(reversed(self.path), 1):
                print(f"Step {i}: {step}")
        else:
            print("No solution exists for this problem")


solver = JugSolver(1, 5, 2)
solver.run()
```
